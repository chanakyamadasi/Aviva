```python
import pandas as pd
import json
from collections import Counter
```

# reading and preparing the data

```python
#reading the json file
data = json.load(open(r"C:\Users\chana\Downloads\input_data (1).json"))
data
```

```python
#converting data into list
abstract=[]
label=[]
signatures=[]
for i in range(len(data)):
    abstract.append(data[i]['abstract']['_value'])
    label.append(data[i]['label']['_value'])
    signatures.append(data[i]['numberOfSignatures'])
print('abstract')
print(abstract[:5])
print('label')
print(label[:5])
print('numberOfSignatures')
print(signatures[:5])
```

# Solution 1

```python
#finding the number of words present in abstract,lenght data
abstract_no_words = [len(words.split()) for words in abstract]
label_no_words = [len(words.split()) for words in label]
#creating an unqiue column primary key
primary_key = ['petiton_id_'+str(i) for i in range(len(data))]
#creating an data frame from the obtained data and converting into csv
solution_df1 = pd.DataFrame({'petition_id':primary_key,'label_length':label_no_words,
                   'abstract_length':abstract_no_words,'num_signatures':signatures})
solution_df1.to_csv(r"C:\Users\chana\Downloads\Solution1.csv")
solution_df1
```

# Solution2

```python
#Combining both abstract, label petation data and creating an data frame
combined = [i+j for i,j in zip(abstract,label)]
combined_df = pd.DataFrame({'combined':combined})
#finding the words which has a lenght 0f five or more
fiverormore_letter_word= combined_df['combined'].str.findall('\w{5,}').str.join(' ')
#extracting the most common words among the filtered words 
common_words=Counter(" ".join(fiverormore_letter_word).split()).most_common(21)
#getting the count of occurance of each word in the row 
rows = [combined_df['combined'].str.count(i).tolist() for i in list(zip(*common_words))[0]]
#appending the 21 common words into a list and then created a dataframe from which a csv is created
columns=[i for i in list(zip(*common_words))[0]]
solution_df2 = pd.DataFrame(rows)  
solution_df2=solution_df2.T
solution_df2.columns=columns
solution_df2.insert(loc=0, column='petition_id', value=primary_key)
solution_df2.to_csv(r"C:\Users\chana\Downloads\Solution2.csv")
solution_df2
```
